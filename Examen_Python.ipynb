```python
                        ##### Exercice 1 ######

# Importation des Librairies 

import matplotlib.pyplot as plt
import numpy as np

# Définition de la fonction f(x) = x^5
def f(x):
    return x**5

# Génération des valeurs de x et y pour la courbe
x_values = np.linspace(0, 1, 100)
y_values = f(x_values)

# Tracé de la courbe
plt.plot(x_values, y_values, label='y = x^5')

# Tracé du carré
plt.plot([0, 1, 1, 0, 0], [0, 0, 1, 1, 0], color='red')

# Representation graphique
plt.xlabel('x')
plt.ylabel('y')
plt.title('Courbe y = x^5 et carré délimité')
plt.legend()
plt.grid(True)

# Affichage 
plt.show()
```

```python
                    ##### Exercice 2  #####


# Création de la classe Curve
class Curve:
    def __init__(self, start, stop, nbr_points=5432):
        self.start = start
        self.stop = stop
        self.nbr_points = nbr_points
        
    # méthode privée qui calcule et renvoie la valeur de la fonction f
    def _f(self, x):
        return x**5
    
    # place les points dans le graphe
    def plot_curve(self):
        x_values = np.linspace(self.start, self.stop, 1000)
        y_values = self._f(x_values)
        plt.plot(x_values, y_values, label='y = x^5')
        plt.xlabel('x')
        plt.ylabel('y')
        plt.title('Courbe y = x^5')
        plt.grid(True)
        
        x_points, y_points = self.generate_points()
        for i in range(len(x_points)):
            if self.is_above_curve(x_points[i], y_points[i]):
                plt.plot(x_points[i], y_points[i], 'bx')  # Point bleu si au-dessus de la courbe
            else:
                plt.plot(x_points[i], y_points[i], 'go')  # Point vert sinon
        
        plt.legend()
        plt.show()
        
    # verifie si un point se trouve au dessus ou en dessous de la courbe
    def is_above_curve(self, x, y):
        return y > self._f(x)
    
    # genere un certain nombre de points
    def generate_points(self):
        x_values = np.random.uniform(self.start, self.stop, self.nbr_points)
        max_y = max(self._f(x_values))
        y_values = np.random.uniform(0, max_y, self.nbr_points)
        return x_values, y_values
    
    # calcule les aires des deux surfaces
    def calculate_area(self):
        x_values, y_values = self.generate_points()
        delta_x = (self.stop - self.start) / self.nbr_points
        area_blue = sum(y_values[self.is_above_curve(x_values, y_values)] * delta_x)
        area_total = (self.stop - self.start) * max(y_values)
        area_green = area_total - area_blue
        return area_green, area_blue

# Exemple d'utilisation de la classe Curve
curve = Curve(0, 1)
curve.plot_curve()
area_green, area_blue = curve.calculate_area()
print("Aire en vert:", area_green)
print("Aire en bleu:", area_blue)


```
